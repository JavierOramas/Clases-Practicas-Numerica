# 6 - Cuando dos más dos no siempre es cuatro. . .(40000créditos)
Dados una aritmética de punto flotante F = (10, 3, −5, 5), y los valores S1 y S2 donde:

$$ S_1 = 1 + \sum_{i = 1}^{1000}0.001 , S_2 = \sum_{i=1}^{1000} 0.001+1$$

explique por qué S1 y S2 no tienen el mismo valor si las operaciones se realizan en la aritmética
de punto flotante F.

## Respuesta:
En $S_1$ la sumatoria se realiza de la forma 1+0.001+0.001+...+0.001 (sumando 0.001 1000 veces al 1) pero como en la representación de esta aritmética, al sumar 1 + 0.001, 0.001 se convierte en 0 debido al truncamiento, por tanto el resultado de $S_1$ va a ser 1, en $S_2$ por el contrario, la sumatoria se "ejecuta" primero (y como estan en el mismo exponente todos los 0.001), dando como resultado que la suma final sea de 1+1, evitando asi el error por truncamiento, lo cual hace que el resultado $S_2$ sea 2

## Respuesta:
todo número x que sea representado por el formato de doble precisión de IEEE 754 (como los float de python) cumple que: $ 1^-499 \leq x \leq 9.9999... * 10^499 $

Si tomamos $r(x) = x^{\frac{1}{2}^{70}}, \forall x \geq 1$ (en una aritmética de punto flotante) se cumple:
$$ 1 \leq r(x) < r(10^500) = 10^{\frac{500}{2}^{70}} $$
$$ = e^{(500*2)^{-70}* \log(10)}  < e^{{10}^{-15}} $$
$$ = 1 +10^{-15} + \frac{1}{2} * 10^{-30} + ... $$
$$ \approx 1 $$

Como el proceso repetitivo de aplicar raíz cuadrada eventualmente convierte el número en 1, el proceso de potenciación al cuadrado de este resultado queda igual, siendo 1 el resultado para las x mayores que 1

Para $ 0 < x < 1 $ tenemos:
$$ x \leq 0.9999999999999999 $$
por lo que la raíz cuadrada va a satisfacer:

$$ \sqrt(x) \leq (1-10^{-16})^{\frac{1}{2}}) 1 - \frac{1}{2} * 10^{-16} - \frac{1}{8}* 10^{-32}$$
$$ = 0.99999999999999994... $$

$$ s(x) \leq s(0.99...9) = (1-10^{-16})^{2^{70}}$$
lo cual es un número menor que el menor representable en la aritmética, por tanto ocurre un underflow, lo que nos da como resultado, el 0
